```python
small_1_df=small_df.loc[df['INSTNM'].isin(['University of Maryland-Baltimore County','University of Maryland-College Park','Loyola University Maryland','Johns Hopkins University'])]

# Finding and displaying University of Maryland, College Park, Loyola College Maryland, Johns Hopskins
```

```python
max_tution_fee=small_1_df["TUITIONFEE_IN"].max()
print("The maximum tution fee of the 4 universities is", max_tution_fee)

min_tution_fee=small_1_df["TUITIONFEE_IN"].min()
print("The minimum tution fee of the 4 universities is", min_tution_fee)

avg_tution_fee=small_1_df["TUITIONFEE_IN"].mean()
print("The average tution fee of the 4 universities is", avg_tution_fee)
# Calculating the min, max, and average in-state tution of the above four institutions.
```
