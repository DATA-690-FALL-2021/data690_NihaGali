```python
df = pd.read_csv("https://raw.githubusercontent.com/NihaRGali/data690_NihaGali/main/project_01/wdi_data%20(3).csv")
df
```

```python
df.drop(columns=["Unnamed: 0"],inplace=True)
df
```

```python
df_2015 = df.query("Year == 2015")
df_2015.sample(10)
```

```python
df_2015 = df_2015.query("indicator in['NY.GDP.PCAP.PP.CD','SP.POP.TOTL','SH.STA.SUIC.P5	','SP.DYN.LE00.IN	']")
df_2015.head()
```

```python
df_pivot = df_2015.pivot(index =['Country Name','Country Code','Region','Income Group','Lending Type'], columns='indicator', values='value').reset_index()
df_pivot.head()
```

```python
df_pivot.info()
```

```python
fig = px.bar(
    df_pivot,
    x="Country Name",
    y="NY.GDP.PCAP.PP.CD",
    title="TOP 10 Countries by 2015 GDP,PPP(Constant 2017 international $",
    labels={"NY.GDP.PCAP.PP.CD":"2015 GDP,PPP(Constant 2017 international $)"}, 
    color="Country Name"
)
fig.update_xaxes(categoryorder="total descending")
fig.show()
```

```python
df1= pd.read_csv("https://raw.githubusercontent.com/NihaRGali/data690_NihaGali/main/project_01/wdi_data%20(5).csv")
```

```python
df_pop = df1.query('indicator == "SP.POP.TOTL"')
df_pop.head(5)
```

```python
df_pivot1 = df_pop.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'],columns = 'indicator', values ='value').reset_index()
df_pivot1.reset_index().head()
```

```python
fig = px.line(
    df_pivot1, 
    x="Year", 
    y="SP.POP.TOTL",
    labels={"SP.POP.TOTL":"Population, female(% of total population)"},
    hover_name='Country Name',
    color='Country Name'
    )
fig.show()
```

```python
df2= pd.read_csv("https://raw.githubusercontent.com/NihaRGali/data690_NihaGali/main/project_01/wdi_data%20(6).csv")
```

```python
df_pop1 = df2.query('indicator == "SP.POP.TOTL"')
df_pop1.head(5)
```

```python
df_pivot2 = df_pop1.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'],columns = 'indicator', values ='value').reset_index()
df_pivot2.reset_index().head()
```
