```python
import pandas as pd
import plotly.express as px
```

```python
df= pd.read_csv("/content/wdi_data.csv")
df.head()
```

```python
df.sample(6)
```

```python
df_2019 = df.query("Year == 2019")

df_2019.sample(10)
```

```python
df_2019 = df_2019.query("indicator in ['SP.DYN.LE00.IN', 'NY.GDP.PCAP.PP.CD','SP.POP.TOTL'] ")
df_2019
```

```python
df_2019 = df_2019[df_2019["indicator"].isin(['SP.DYN.LE00.IN', 'NY.GDP.PCAP.PP.CD','SP.POP.TOTL'])]
df_2019
```

```python
df_2019.drop(columns = ['Unnamed : 0','Year'], inplace = True)
df_2019.head()
```

```python

df_2019.drop(columns = ['Unnamed : 0 ', 'Year'])
```

```python
df_pivot =  df_2019.pivot_table(index=['Country Code', 'Country Name', 'Region','Income Group','Lending Type'], columns = 'indicator', values = 'value').reset_index()
df_pivot.head()
```

```python
fig = px.scatter(
    df_pivot,
    x="NY.GDP.PCAP.PP.CD",
    y="SP.DYN.LE00.IN",
    title = "Health vs wealth",
    labels = {"SP.DYN.LE00.IN":"Life Expectancy", "NY.GDP.PCAP.PP.CD":"GDP per Capita"},
    color ="Region",
    hover_name="Country Name",
    text="Country Name",
   
)

fig.update_traces(textposition='top center')

fig.show()
```
